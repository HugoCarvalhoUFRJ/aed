## Gráficos para variáveis quantitativas

_Gráfico de barras_ ou _gráfico de dispersão unidimensional_

Exemplo com a variável "Número de filhos"

```r
barplot(table(empresa$Filhos)) # Melhorando novamente
```

```r
barplot(table(empresa$Filhos),
        ylab = "Frequência",
        col = "darkgrey",
        main = "Gráfico em barras para a variável Z: Número de filhos.",
        border = NA)
```

```r
stripchart(empresa$Filhos,
           method = "stack",
           offset = 1,
           pch = 19,
           col = "darkblue",
           ylim = c(0,7),
           ylab = "Frequência",
           xlab = "Número de filhos",
           cex = 1)
```

```r
table(empresa$Filhos)
```

```r
plot(table(empresa$Filhos))
```

```r
plot(table(empresa$Filhos),
     type = "p",
     col = "darkblue",
     pch = 19,
     ylab = "Frequência",
     xlab = "Número de filhos")
```

Exemplo com a variável "Salário"

```r
barplot(table(cut(empresa$Salario, breaks = seq(4, 24, by = 4), right = FALSE)),
        ylab = "Frequência",
        col = "darkgrey",
        border = NA)
```

Histograma

```r
hist(empresa$Salario,
     breaks = seq(4, 24, by = 4),
     right = FALSE,
     probability = TRUE,
     ylab = "Densidade de frequência",
     xlab = "Salário",
     main = "Histograma dos salários",
     xlim = c(0, 24),
     xaxp = c(0, 24, 6),
     ylim = c(0, 0.1))
```

```r
hist(empresa$Filhos,
     breaks = seq(-.5, 5.5, by = 1),
     right = FALSE,
     ylab = "Densidade de frequência",
     xlab = "Número de filhos",
     main = "Histograma do número de filhos",
     ylim = c(0, 8))
```

Histograma simulado

```r
n = 1000000

pop_x <- seq(-6, 6, length.out = 1000)
x <- rnorm(n)
```

```r
delta = 0.1

breaks = seq(-6, 6, delta)
hist(x, breaks, prob = TRUE)
lines(pop_x, dnorm(pop_x), col = "red")
```
