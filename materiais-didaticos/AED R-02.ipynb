Várias possibilidades. Todas variação de _gráfico em barras_ e _gráfico em setores_ (o famoso _gráfico de pizza_).

Exemplo com a variável "Grau de Instrução"

```r
table(empresa$GrauInstrucao)
```

```r
barplot(table(empresa$GrauInstrucao)) # Feio, melhorar!
```

```r
barplot(table(empresa$GrauInstrucao),
        ylab = "Frequência",
        names.arg = c("Fundamental", "Médio", "Superior"),
        col = "darkgrey",
        main = "Gráfico em barras para a variável Y: Grau de Instrução",
        border = NA,
        axes = TRUE,
        ylim = c(0,20))
```
