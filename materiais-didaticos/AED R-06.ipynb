### Meus dados são normais?!

```r
dados = rbinom(100, 1, 0.5)
```

```r
table(dados)
```

$$\begin{align*} \mathcal{H}_0 &= \text{ dados normais} \\ \mathcal{H}_1 &= \text{ dados não normais} \end{align*}$$

```r
ks.test(dados, "pnorm")
```

Como esse teste é feito?

```r
plot(ecdf(dados)) # empirical cumulative distribution function
```

```r
plot(pnorm, from = -6, to = 6)
```

```r
plot(ecdf(dados))
par(new = TRUE)
plot(pnorm, from = -6, to = 6)
```

Com um cenário "real".

```r
empresa <- read.table("empresa.csv", dec = ",", sep = ";", h = T)
```

```r
plot(ecdf(empresa$IdadeAnos))
```

```r
plot(ecdf((empresa$IdadeAnos - mean(empresa$IdadeAnos))/sd(empresa$IdadeAnos)))
par(new = TRUE)
plot(pnorm, from = -6, to = 6)
```

```r
ks.test((empresa$IdadeAnos - mean(empresa$IdadeAnos))/sd(empresa$IdadeAnos), "pnorm")
```

```r
hist((empresa$IdadeAnos - mean(empresa$IdadeAnos))/sd(empresa$IdadeAnos), breaks = seq(-4, 4, by = 0.5))
par(new = TRUE)
curve(dnorm, from = -4, to = 4)
```
